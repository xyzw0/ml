```python
import pandas as pd
from IPython.display import Video

df = pd.read_csv('/content/hypo.csv')

display(df.head())
```

```python
print(df.shape)
```

```python
for i in df.columns:
  print(f'{i} : {df[i].unique()}')
```

```python
result=df['Buy'].values

attributes=df.drop('Buy',axis=1).values

Hypothesis=['0']*attributes.shape[1]

print(f'Initial Hypothesis is : {Hypothesis}')
```

```python
for i in range(attributes.shape[0]):
  if result[i]=='Yes':
    for j in range(attributes.shape[1]):
      if Hypothesis[j]=='0':
        Hypothesis[j]=attributes[i][j]
      elif Hypothesis[j]!=attributes[i][j]:
        Hypothesis[j]='?'

  print(f'Step-{i} : {Hypothesis}')

print(f'Final Hypothesis : {Hypothesis}')
```
