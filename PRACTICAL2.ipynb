```python
#Exporting data to new file
PoliticianData = {
    'Name': ['Narendra Modi', 'Rahul Gandhi', 'Mamata Banerjee', 'Arvind Kejriwal'],
    'Party': ['BJP', 'INC', 'AITC', 'AAP'],
    'State': ['Gujarat', 'Kerala', 'West Bengal', 'Delhi'],
    'Age': [73, 53, 69, 55]
}

df_politicians = pd.DataFrame(PoliticianData)

df_politicians.to_csv('PoliticianData.csv', index=False)

df_politicians_read = pd.read_csv('PoliticianData.csv')
display(df_politicians_read)
```

```python
# Conditional modification of rows
for index, row in df_politicians.iterrows():
    if row['Age'] >= 60:
        df_politicians.at[index, 'SeniorCitizen'] = 'Yes'
    else:
        df_politicians.at[index, 'SeniorCitizen'] = 'No'

display(df_politicians)
```

```python
# Conditional modification of columns
for column_name, column_data in df_politicians.items():
    if column_name == 'Age':
        df_politicians[column_name] = column_data + 2

display(df_politicians)
```

```python
#Use Pandas for masking data and reading in Boolean format

df = pd.DataFrame({
    'A': [3, 8, 15, 22, 37],
    'B': [14, 28, 35, 7, 24],
    'C': [50, 60, 70, 80, 90]
})

print("Original DataFrame:")
print(df)

# Create a mask with multiple conditions: (A < 20) & (B > 10)
mask = (df['A'] < 20) & (df['B'] > 10)
print("\nMask for (A < 20) & (B > 10):")
print(mask)

# Apply the combined mask
filtered_df = df[mask]
print("\nFiltered DataFrame (A < 20 and B > 10):")
print(filtered_df)

# Boolean format for condition A > 20
boolean_format_A = df['A'] > 20
print("\nBoolean format for condition (A > 20):")
print(boolean_format_A)

# Boolean format for condition B != 24
boolean_format_B = df['B'] != 24
print("\nBoolean format for condition (B != 24):")
print(boolean_format_B)
```

```python
#Importing matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Simple line chart
x = np.array([1, 2, 3, 4, 5])
y = x*3

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Line Chart of Y = 3X')

plt.plot(x, y)
plt.show()
```

```python
#Corelation chart
y = pd.Series([2, 3, 5, 6, 4, 7, 5])
x = pd.Series([1, 2, 3, 4, 5, 6, 7])

correlation = y.corr(x)
print("Correlation:", correlation)

plt.title('Correlation Chart')
plt.scatter(x, y)

plt.plot(np.unique(x),
         np.poly1d(np.polyfit(x, y, 1))
         (np.unique(x)), color='red')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
#Histogram
data = np.random.uniform(low=-3, high=3, size=1000)

plt.hist(data, bins=30, color='orange', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with Uniform Distribution')

plt.show()
```

```python
#Plotting of Multivariate data
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

def func(x, y):
   return 3 * x + 4 * y - 2 + np.random.randn(50)

x, y = np.random.randn(2, 50)
y *= 50
z = func(x, y)

fig, ax = plt.subplots()

s = ax.scatter(x, y, c=z, s=100, marker='o', cmap='viridis')

fig.colorbar(s)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multivariate Line Plot')
plt.show()
```

```python
#Plotting of Pie chart
teams = ['Manchester United', 'Liverpool', 'Chelsea', 'Arsenal', 'Manchester City', 'Tottenham Hotspur']
wins = [30, 28, 26, 24, 32, 22]

fig = plt.figure(figsize=(10, 7))
plt.pie(wins, labels=teams)

plt.title('Football Team Wins Distribution')
plt.show()
```
